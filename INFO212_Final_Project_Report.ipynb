```python
import pandas as pd

def dfDisplayInfo(df):
    """Print DataFrame info"""

    obs = df.shape[0]
    info = pd.DataFrame({
        'Types': df.dtypes,
        'Counts': df.count(),
        'Nulls': df.isnull().sum(),
        'Missing Ratio': (df.isnull().sum() / obs) * 100,
        'Unique Values': df.nunique(),
        'Skewness': df.skew()
    })

    print(f"There are {df.shape[1]} columns and {df.shape[0]} rows in the dataset.")
    print(f"There are {df[df.duplicated()].shape[0]} duplicated rows in the dataset.")
    print(f"There are {df['Id'].nunique()} unique values in the 'Id' column.")
    print('___________________________\nData types:\n', info['Types'].value_counts())
    print('___________________________')
    print('Descriptive statistics of the numeric features in the dataset:')
    print(df.describe())

    return info
```

```python
# Calculate descriptive statistics
mean_values = analysis_df1.mean()
median_values = analysis_df1.median()
std_values = analysis_df1.std()
min_values = analysis_df1.min()
max_values = analysis_df1.max()
percentiles = analysis_df1.describe(percentiles=[0.25, 0.5, 0.75]).loc[['25%', '50%', '75%']]

# Create a DataFrame to display the statistics
statistics_df = pd.DataFrame({'Mean': mean_values,
                              'Median': median_values,
                              'Std': std_values,
                              'Min': min_values,
                              'Max': max_values,
                              '25th percentile': percentiles.loc['25%'],
                              '50th percentile': percentiles.loc['50%'],
                              '75th percentile': percentiles.loc['75%']})

pd.set_option('display.float_format', '{:.2f}'.format)
print(statistics_df)
```

```python
#Plot the average step counts for each of the users

average_steps = analysis_df1.groupby('Id')['TotalSteps'].mean()

plt.figure(figsize=(10, 6))
average_steps.plot(kind='bar')
plt.xlabel('User ID')
plt.ylabel('Average Step Count')
plt.title('Average Daily Step Counts for Each User')
plt.xticks(rotation=90)
plt.show()
```

```python
#Create side-by-side box plots for the steps of each user

plt.figure(figsize=(10, 6))

# Create a box plot for each user
sns.boxplot(data=analysis_df1, x='Id', y='TotalSteps')
plt.xlabel('User ID')
plt.ylabel('Step Count')
plt.title('Step Counts for Each User')

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Plot the percentage of users who reached the step count goal (>=10000 steps) over time
daily_steps = analysis_df1.groupby(['Id', 'ActivityDate'])['TotalSteps'].sum().reset_index()
daily_steps['ReachedGoal'] = daily_steps['TotalSteps'] >= 10000
percentage_reached_goal = daily_steps.groupby('ActivityDate')['ReachedGoal'].mean() * 100

plt.figure(figsize=(10, 6))
plt.plot(percentage_reached_goal.index, percentage_reached_goal, marker='o')
plt.xlabel('Date')
plt.ylabel('Percentage of Users Reached Goal')
plt.title('Percentage of Users Reached 10,000 Steps Goal')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
#Compare users' steps to intensities

# Calculate average steps for each user
average_steps = analysis_df1.groupby('Id')['TotalSteps'].mean()

# Define the step count thresholds for each category
thresholds = {
    'Sedentary': (0, 5000),
    'Lightly': (5000, 7499),
    'Fairly': (7500, 9999),
    'VeryActive': (10000, np.inf)
}

# Categorize each user based on average steps
user_categories = {}

for user_id, avg_steps in average_steps.items():
    for category, (lower, upper) in thresholds.items():
        if lower <= avg_steps <= upper:
            user_categories[user_id] = category
            break

# Print the category for each user
for user_id, category in user_categories.items():
    print(f"User {user_id} is categorized as {category}")

category_counts = {category: list(user_categories.values()).count(category) for category in set(user_categories.values())}
print(category_counts)

# Count the number of users in each category
category_counts = {category: list(user_categories.values()).count(category) for category in set(user_categories.values())}

# Create a pie chart
plt.pie(category_counts.values(), labels=category_counts.keys(), autopct='%1.1f%%')

# Add a title
plt.title("User Distribution by Step Categories")

# Display the chart
plt.show()
```

```python
#Linear regression line plot comparing the users' average daily steps to the users' average daily calories

from sklearn.linear_model import LinearRegression

# Calculate average daily steps and average daily calories burned for each user
average_steps = analysis_df1.groupby('Id')['TotalSteps'].mean()
average_calories = analysis_df1.groupby('Id')['Calories'].mean()

# Create a scatter plot
plt.scatter(average_steps, average_calories)

# Fit a linear regression model
model = LinearRegression()
X = average_steps.values.reshape(-1, 1)
y = average_calories.values.reshape(-1, 1)
model.fit(X, y)

# Generate predicted values
y_pred = model.predict(X)

# Plot the linear regression line
plt.plot(X, y_pred, color='red')

# Set plot labels and title
plt.xlabel('Average Daily Steps')
plt.ylabel('Average Daily Calories Burned')
plt.title('Linear Regression - Average Daily Steps vs Calories Burned')

# Display the plot
plt.show()
```

```python
#correlation of steps and calories
steps_vs_cals = pd.merge(average_steps, average_calories, on = 'Id')

steps_vs_cals.corr()
```

```python
from scipy.stats import linregress

# Filter dataframe for the specific users
user1_df = analysis_df1[analysis_df1['Id'] == 8877689391]
user2_df = analysis_df1[analysis_df1['Id'] == 1927972279]

# Create separate dataframes for each user with required columns
user1_data = user1_df[['TotalSteps', 'Calories']].dropna()
user2_data = user2_df[['TotalSteps', 'Calories']].dropna()

# Perform linear regression for each user
slope1, intercept1, _, _, _ = linregress(user1_data['TotalSteps'], user1_data['Calories'])
slope2, intercept2, _, _, _ = linregress(user2_data['TotalSteps'], user2_data['Calories'])

# Create scatter plot for each user
plt.scatter(user1_data['TotalSteps'], user1_data['Calories'], label='User 1')
plt.scatter(user2_data['TotalSteps'], user2_data['Calories'], label='User 2')

# Plot linear regression line for each user
x_values = np.linspace(0, max(analysis_df1['TotalSteps']), 100)
plt.plot(x_values, slope1 * x_values + intercept1, color='red', label='User 1 Regression Line')
plt.plot(x_values, slope2 * x_values + intercept2, color='green', label='User 2 Regression Line')

# Set plot labels and title
plt.xlabel('Daily Steps')
plt.ylabel('Daily Calories Burned')
plt.title('Daily Steps vs Daily Calories Burned')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
analysis_df2 = pd.read_csv('analysis_df2.csv')
```

```python
#Create user sleep categories

# Calculate the average total steps and average total minutes asleep per user
average_steps = analysis_df2.groupby('Id')['TotalSteps'].mean()
average_minutes_asleep = analysis_df2.groupby('Id')['TotalMinutesAsleep'].mean()

# Create a DataFrame with the average values
average_sleep_data = pd.DataFrame({'AverageTotalSteps': average_steps, 'AverageTotalMinutesAsleep': average_minutes_asleep})

# Calculate the average total minutes asleep for each user
average_minutes_asleep = analysis_df2.groupby('Id')['TotalMinutesAsleep'].mean()

# Categorize the average total minutes asleep
categories = []
for minutes in average_minutes_asleep:
    if minutes > 540:  # More than 9 hours (9 hours * 60 minutes = 540 minutes)
        categories.append('Long')
    elif minutes >= 420:  # Between 7 and 9 hours (7 hours * 60 minutes = 420 minutes)
        categories.append('Moderate')
    else:  # Less than 7 hours
        categories.append('Light')

# Create a new column in the dataframe for the categories
average_sleep_data['SleepCategory'] = categories
```

```python
#Pie chart visualization of average user sleep time

# Count the number of users in each sleep category
sleep_counts = average_sleep_data['SleepCategory'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sleep_counts, labels=sleep_counts.index, autopct='%1.1f%%', startangle=90)

# Add a title and adjust the aspect ratio for a circular pie
plt.title('Distribution of Average Sleep Categories')
plt.axis('equal')

# Add a legend
legend_labels = ['Light: < 7 hours', 'Moderate: 7-9 hours', 'Long: > 9 hours']
plt.legend(legend_labels)

# Display the chart
plt.show()
```

```python
#Linear Regression comparing users' average daily steps to the users' average total minutes asleep

from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

# Fit the model using average steps and average total minutes asleep
model.fit(average_sleep_data[['AverageTotalSteps']], average_sleep_data[['AverageTotalMinutesAsleep']])

# Plot the linear regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='AverageTotalSteps', y='AverageTotalMinutesAsleep', data=average_sleep_data)

# Add labels and title to the plot
plt.xlabel('Average Steps')
plt.ylabel('Average Total Minutes Asleep')
plt.title('Correlation between Average Steps and Average Total Minutes Asleep')

# Show the plot
plt.show()
```

```python
#Linear Regression comparing users' average daily calories to the users' average total minutes asleep

average_calories = analysis_df2.groupby('Id')['Calories'].mean()
average_sleep_data = pd.DataFrame({'AverageTotalSteps': average_steps, 'AverageTotalMinutesAsleep': average_minutes_asleep, 'AverageTotalCalories' : average_calories, 'SleepCategory' : categories})

# Create a linear regression model
model = LinearRegression()

# Fit the model using average calories and average total minutes asleep
model.fit(average_sleep_data[['AverageTotalCalories']], average_sleep_data[['AverageTotalMinutesAsleep']])

# Plot the linear regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='AverageTotalCalories', y='AverageTotalMinutesAsleep', data=average_sleep_data)

# Add labels and title to the plot
plt.xlabel('Average Calories Burned')
plt.ylabel('Average Total Minutes Asleep')
plt.title('Correlation between Average Calories Burned and Average Total Minutes Asleep')

# Show the plot
plt.show()
```
